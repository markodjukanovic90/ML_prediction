```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data_2012 = pd.read_csv("weather_2012.csv") 
data_2012 = data_2012.drop(columns = ["Unnamed: 0"], axis = 1)
```

```python
data_2012.info()
```

```python
data_2012 = data_2012.drop(columns =["year"], axis = 1)
```

```python
data_2012
```

```python
influenza = pd.read_csv("influenza.csv")
influenza = influenza.drop(columns = ["BEZEICHNUNG", "ID" ], axis = 1)
```

```python
influenza_2012 = influenza[ influenza["Jahr"] == 2012 ]
```

```python
influenza_2012
```

```python
data_2012 = data_2012.drop(columns = [ "month", "day" ], axis = 1)
```

```python
dataX = []
data_2012
for i in range(0, len(data_2012)-7, 7):
    mean_i = data_2012[i : i + 7  ].mean()
    mean_i = mean_i.reset_index()
    arr = ((list)( mean_i.iloc[:, [1]].to_numpy().flatten()) )
    #mean_x = mean_i[:, [1]]
    #mean_x
    dataX.append(arr)
#dataX
data_2012_avg = pd.DataFrame( dataX )
influenza_2012 = influenza_2012.reset_index()
influenca_week = influenza_2012[[ "Neuerkrankungen pro Woche" ]]
influenca_week ## should be fixed
data_2012_avg_influenza = pd.concat([ influenca_week, data_2012_avg ], axis = 1)
```

```python
influenza_2012
```

```python
for i in range(data_2012_avg_influenza.shape[0]):
    if i <= 14: 
        data_2012_avg_influenza.loc[i, ["Neuerkrankungen pro Woche"]] = influenza_2012.loc[i, ["Neuerkrankungen pro Woche"]]
    else:
        data_2012_avg_influenza.loc[i, ["Neuerkrankungen pro Woche"]] =  0

# 14 ==> 25:
start2 = 14
for i in range(0, influenza_2012.shape[0] - start2  ):
    data_2012_avg_influenza.loc[ data_2012_avg_influenza.shape[0] - influenza_2012.shape[0] + start2 + i, ["Neuerkrankungen pro Woche"]] = influenza_2012.loc[start2 + i, ["Neuerkrankungen pro Woche"]]
    
data_2012_avg_influenza    
```

```python
data_2012_avg_influenza
import seaborn as sns 

x_vals = np.arange(len(data_2012_avg_influenza["Neuerkrankungen pro Woche"]))
y_vals = data_2012_avg_influenza["Neuerkrankungen pro Woche"].values
y_vals = y_vals.astype("int32")

plt.figure(figsize=(9, 9))
plt.plot(x_vals ,  y_vals)
plt.xticks([0, 9, 20, 30, 48 ], ["Jan", "Mar", "May", "Aug", "Nov" ])
plt.ylabel("#Infected (avg. per week)")
plt.show()
```

```python
plt.figure(figsize=(10, 10))
sns.heatmap(data = data_2012_avg_influenza.corr(), annot = True)
```

```python
## PCA ==> 10 features (PCA):
from sklearn.decomposition import PCA
pca = PCA(n_components = 8) 
```

```python
Y = data_2012_avg_influenza[["Neuerkrankungen pro Woche"]]
X = data_2012_avg_influenza.drop(columns = ["Neuerkrankungen pro Woche"], axis = 1)
```

```python
X.info()
```

```python
pca.fit(X)
X_pca = pca.transform(X)
X_pca # transformed
```

```python
## prediction ==> forecasting: 
from sklearn.neural_network import MLPRegressor
#solver='adam'
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_pca, Y, test_size = 0.3, random_state = 3)

mlpr = MLPRegressor( solver = 'lbfgs', max_iter = 3000   )
mlpr.fit(X_train, y_train)

### KnearesrRegressor 
from sklearn.neighbors import KNeighborsRegressor
knr = KNeighborsRegressor()
knr.fit(X_train, y_train) 
###RandomForestRegressor:
from sklearn.ensemble import RandomForestRegressor
rfr = RandomForestRegressor(n_estimators = 10)
rfr.fit(X_train, y_train)
```

```python
y_pred = mlpr.predict(X_test)
#y_pred_knr = knr.predict(X_test)
y_pred_rfr = rfr.predict(X_test)
```

```python
from sklearn.metrics import r2_score
print("score: ", r2_score(y_pred, y_test)) # 0.11 (weak)
print("score of rfr: ", r2_score(y_pred_rfr, y_test)) #  -0.07511064195291595 (weak)
```

```python
plt.figure(figsize=(9, 9))
plt.plot(x_vals ,  y_vals)
plt.plot([ i for i in range(len(y_pred))] ,  y_pred)

plt.xticks([0, 9, 20, 30, 48 ], ["Jan", "Mar", "May", "Aug", "Nov" ])
plt.ylabel("#Infected (avg. per week)")
plt.show()
```

```python
### enhanced models for forecasting (arima models):
from statsmodels.tsa.arima_model import ARIMA
from statsmodels.tsa.statespace.sarimax import SARIMAX
```

```python
y_trainD = pd.DataFrame(y_train)
Y.info()
ylist =  (Y["Neuerkrankungen pro Woche"].to_numpy().flatten())
ylist = ylist.astype("int32")
ylist
arima_model = SARIMAX(ylist, order = (2,1,1), seasonal_order = (4,0,3,12))
arima_result = arima_model.fit()
arima_result.summary()
```

```python
## arima predict:
arima_pred = arima_result.predict(start = 0, end = len(ylist)-1, typ="levels")#.rename("ARIMA Predictions")
arima_pred
```

```python
plt.figure(figsize=(9, 9))
plt.plot(x_vals ,  y_vals)
plt.plot([ i for i in range(len(arima_pred))] ,  arima_pred)

plt.xticks([0, 9, 20, 30, 48 ], ["Jan", "Mar", "May", "Aug", "Nov" ])
plt.ylabel("#Infected (avg. per week)")
plt.show()
print("Error of ARIMA model forecasting: ", r2_score(arima_pred, y_vals)) # 0.86 precision...
```
